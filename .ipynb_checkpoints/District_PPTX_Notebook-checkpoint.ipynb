# Build Slides

This is the main section where the slide are built.  It takes a few minutes to run for two slides per district.

```python
# Current File Locations

Current_template = 'District TCAP Results AB.pptx'
file_name = 'District_TCAP_Results_11_smooth_new.pptx'  #.format(year = datetime.datetime.now().year, month = datetime.datetime.now().month, day = datetime.datetime.now().day)

# Save a new copy of the template
prs = pptx.Presentation(Current_template)
prs.save(file_name)
prs = pptx.Presentation(file_name)

## These are the positions of the master slides with placeholders
ELA3_num = 2
Math_num = 3
ELA4_num = 4

# Placeholder positions
title_placeholder = 0
line_placeholder = 12
lines_placeholder = 13
donut_placeholder = 11
text_placeholder = 15
model_placeholder = 25

## Load the layouts of the master slides
ELA3_slide_master = prs.slide_layouts[ELA3_num]
Math_slide_master = prs.slide_layouts[Math_num]
ELA4_slide_master = prs.slide_layouts[ELA4_num]

district_subset = df.sort_values('system_name').system.unique()
# district_subset = [10,171,51,985,11,792,190,340,95,690,581,960,963,964] #These are the test-case districts

## Formatting and display settings
smooth = True  ## whether to have smooth lines on the line charts
participation_offset = 55  ## This value determines how much change the donut chart will show

## Data Construction and Formatting Functions for the longitudinal line charts
def get_line_data(sub, grade, LEA):    
    '''corrects for data anomalies'''
    subpop = LEA.loc[LEA.subgroup.isin([sub]) & (LEA.grade == grade)][['year','pct_met_exceeded_supp_num']].set_index('year')['pct_met_exceeded_supp_num'] # Old: ['pct_on_mastered_supp_text']
    subpop = tuple([subpop[y] if y in subpop.index else None for y in yr_idx])
    subpop = tuple(None if isinstance(i, float) and math.isnan(i) else i for i in subpop)
    subpop = tuple(None if pd.isnull(i) or (i == None) else i for i in subpop)
    return subpop

def format_line_chart(chart):
    '''Quick function to format all line graphs the same.'''
    chart.has_legend = False
    chart.category_axis.tick_labels.font.color.rgb = axes_color
    chart.category_axis.tick_labels.font.size = axes_font_size
    chart.value_axis.tick_labels.font.color.rgb = axes_color
    chart.value_axis.tick_labels.font.size = axes_font_size
    chart.value_axis.major_gridlines.format.line.color.rgb = axes_color
    chart.category_axis.major_tick_mark = XL_TICK_MARK.NONE
    chart.value_axis.format.line.fill.background()    
#     chart.element[1][4].set('val','span')  #code to get non-connecting dots to connect
    chart.value_axis.maximum_scale = 100.0  #       Maximum
    chart.value_axis.minimum_scale = 0.0  #       minimum    
     
    return chart

# for each district
for LEA_n in district_subset:

    LEA = df[(df.system == LEA_n)]
    
    ## if they have scores for 2022 
    if 2022 in list(LEA[LEA.subgroup == 'All Students'].year.unique()):

        ## Make two slides
        ELA3 = prs.slides.add_slide(ELA3_slide_master)
        ELA4 = prs.slides.add_slide(ELA4_slide_master)
        Math = prs.slides.add_slide(Math_slide_master)

        ## and Connect them to data
        slides = [ELA3,ELA4,Math]
        grades = [3, 4, 7]
        state_results = [TN_3ELA, TN_4ELA, TN_7Math]
        
        ## populate each of the slides with
        for s, g, TN in zip(slides, grades,state_results):

            ## District Name
            title = s.placeholders[title_placeholder]
            title.text = df[df.system == LEA_n].system_name.iloc[-1] ## Take the last recorded name (as years are ascending order)

            ## Text of 2021 operating Model from dist_models data frame
#             modelbox = s.placeholders[model_placeholder]
#             modelbox.text = dist_models["Most Common In Person"][LEA_n] 

            ## Text Percent for participation
            participation = part_df.LEA_participation[LEA_n]
            part = s.placeholders[text_placeholder]  # idx key, not position
            part.text = str(int(participation)) + "%"    
            
            ## Donut Graph of participation
            donut = s.placeholders[donut_placeholder]  # idx key, not position
            part_data = ChartData()
            part_data.categories = ['Participated', 'Not']
            part_data.add_series('Series 1', (participation - participation_offset, 100-participation))  ## Set the data in a series

            graphic_frame = donut.insert_chart(XL_CHART_TYPE.DOUGHNUT, part_data)
            part_chart = graphic_frame.chart
            part_chart.has_legend = False
            part_chart.has_title = False

            ### custom formatting
            for p in range(len(part_data.categories)):
                part_chart.series[0].points[p].format.line.color.rgb = donut_borders
                part_chart.series[0].points[p].format.line.width = donut_line_width
                part_chart.series[0].points[p].format.fill.solid()

            part_chart.series[0].points[0].format.fill.fore_color.rgb = donut_blue
            part_chart.series[0].points[1].format.fill.fore_color.rgb = donut_gray   
            part_chart.element[2][1][1][4].set('val','65') ## This is for the hole size in the donut chart
            
            ## Proficiency Placeholders for each student group
            for sub in focus_groups:
                if LEA.loc[LEA.subgroup.isin([sub]) & (LEA.grade == g) & (LEA.year == 2022) & pd.notnull(df.pct_met_exceeded)].shape[0] > 0:
                    legend_labels = s.placeholders[group_params[sub]['pholder']]
                    legend_labels.text = LEA.loc[LEA.subgroup.isin([sub]) & (LEA.grade == g) & (LEA.year == 2022) & pd.notnull(df.pct_met_exceeded)]['pct_met_exceeded'].iloc[0]  # Old: ['pct_on_mastered_supp_text']
                else:
                    legend_labels = s.placeholders[group_params[sub]['pholder']]
                    legend_labels.text = "--"
            legend_labels = s.placeholders[group_params['All TN Students']['pholder']]
            legend_labels.text = str(round(TN[2022],1)) +"%"  ## Old: ['pct_on_mastered']

            ### Line Chart 1: Line Graph state and district average
            line_chart = s.placeholders[line_placeholder]  # idx key, not position
            line_chart_data = ChartData()
            line_chart_data.categories = [str(s) for s in yr_idx] #List
            
            sub = 'All Students'
            line_chart_data.add_series(sub, get_line_data(sub = sub, grade = g, LEA = LEA))
            line_chart_data.add_series('All TN Students',tuple(TN))

            graphic_frame = line_chart.insert_chart(XL_CHART_TYPE.LINE, line_chart_data)
            chart = graphic_frame.chart

            ## forrmat each line series
            for i in (0,-1):
                chart.series[i].format.line.color.rgb = chart_colors[i]
                chart.series[i].format.line.width = chart_lines[i]
                chart.series[i].smooth = smooth
            chart.series[-1].format.line.dash_style = all_tn_line_style
            
            chart = format_line_chart(chart)

            ### Line Chart 2: Line Graph Student Groups
            line_chart2 = s.placeholders[lines_placeholder]  # idx key, not position
            line_chart_data2 = ChartData()
            line_chart_data2.categories = [str(s) for s in yr_idx] #List

            #### Add each of the data sets (except the All students group)
            for k, sub in enumerate(focus_groups[1:]):
                line_chart_data2.add_series(sub, get_line_data(sub = sub, grade = g, LEA = LEA))

            graphic_frame2 = line_chart2.insert_chart(XL_CHART_TYPE.LINE, line_chart_data2)
            chart2 = graphic_frame2.chart

            for i, sub in enumerate(focus_groups[1:]):
                chart2.series[i].format.line.color.rgb = group_params[sub]['color']
                chart2.series[i].format.line.width = group_params[sub]['line_width']
                chart2.series[i].smooth = smooth    

            ## Standardize custom formatting for both Line Charts
            chart2 = format_line_chart(chart2)
            
prs.save(file_name)
```

```python
subpops = ['average_daily_membership', 'white_pct', 'african_american_pct',
       'hispanic_pct', 'asian_pct', 'native_american_pct','hawaiian_pacisld_pct', 'english_learner_pct',
       'students_with_disabilities_pct', 'economically_disadvantaged_pct','st_per_grade']

grade_thresh = np.array([0.1,1.1,2.1,3.1,4.1])
letter_grade = ['F','D','C','B','A']
letter_numeric_equivalent = [0,1,2,3,4]


def make_id(s):
    '''Takes the input of a dataframe with columns district # and School #
    and returns a unique identifier by school'''
    return s[0]*100000 +s[1]  

def low_grades(s):
    '''Takes the grades_served variable and returns the low grade as an int'''
    s = str(s)
    if len(s) > 6:
        if re.split(r'\W+',s)[-2] in ['PK','K']:
            return 1-len(re.split(r'\W+',s)[1])
        return int(re.split(r'\W+',s)[1])
    else: return
    
def high_grades(s):
    '''Takes the grades_served variable and returns the high grade as an int'''
    s = str(s)
    if len(s) > 6:        
        return int(re.split(r'\W+',s)[-1])
    else: return


def grade_labels(s):
    '''Converts a letter grade given numeric score into a categorical letter grade'''
    try: 
        pos = np.searchsorted(grade_thresh,s)
        return letter_grade[pos]
    except:
        return
    
    
def stpergrade(s):
    '''Takes the input of a dataframe with columns average_daily_membership and grade_span
    to get avg number of students per grade'''
    if s[1] > 0:
        return s[0]/s[1]
    else: return s[0]
    
    
def score_choice(s):
    '''Takes the inputs of a dataframe with columns metric, score, score_abs, and score_target
    to get the categorical variable indicating which pathway was used'''
    if s[0] == None:
        return
    if s[1] == s[2]:
        return 'tie'
    elif s[0] == s[1]:
        return 'abs'
    else:
        return 'AMO'

def score_diff(s):
    '''Takes the inputs of a dataframe with columns AMO_target and metric to find the 
    magnitude of the difference for the goal'''
    if s[0] == None:
        return
    else:
        return s[1] -s[0]
    
def score_diff2(s):
    '''Takes the inputs of a dataframe with columns AMO_target and metric to find the 
    ratio magnitude of the difference for the goal'''
    if s[0] == None:
        return
    else:
        return (s[1]+1) / (s[0]+1)
```
